```
covid_df = spark.read.csv("/databricks-datasets/COVID/USAFacts/*.csv", header=True, inferSchema=True)
covid_df.limit(10).display()
```

```
from pyspark.sql.functions import input_file_name
covid_df = covid_df.withColumn("file_name", input_file_name())
covid_df.limit(10).display()
```

```
# lets find How to Get no of rows from each file
covid_df.groupBy("file_name").count().display()
print("Total count - ", covid_df.count())
```
